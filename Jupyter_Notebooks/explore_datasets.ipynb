# Explore the population dataset

```python
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
```

```python
df = pd.read_csv('cleaned_df.csv')
```

```python
# First five rows
df.head()
```

```python
# Summary statistics
df.describe()
```

```python
# Data types
df.info()
```

```python
# Identify Missing Values
print(df.isnull().sum())
```

```python
# Z-Score
z_scores = zscore(df.select_dtypes(include='number'))
outliers = (abs(z_scores) > 3).sum(axis=0)
print(outliers)
```

```python
# Value counts for 'Year' column
year_counts = df['Year'].value_counts()
print("Year Counts:\n", year_counts)
```

```python
# Bar plot for average population density by year
avg_density_by_year = df.groupby('Year')['Population density'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='Population density', data=avg_density_by_year)
plt.title('Average Population Density by Year')
plt.xlabel('Year')
plt.ylabel('Population Density')
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot histograms for numeric columns
numeric_cols = ['Population aged 0 to 14 years old (percentage)', 
                'Population aged 60+ years old (percentage)', 
                'Population density']

plt.figure(figsize=(12, 10))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(col)

plt.tight_layout()
plt.show()
```

```python
# Scatter plot between 'Population mid-year estimates (millions)' and 'Population density'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Population mid-year estimates (millions)', y='Population density', data=df)
plt.title('Scatter Plot: Population vs Density')
plt.xlabel('Population (millions)')
plt.ylabel('Population density')
plt.show()
```

```python
# Box plot for 'Population mid-year estimates (millions)' by year
plt.figure(figsize=(10, 6))
sns.boxplot(x='Year', y='Population mid-year estimates (millions)', data=df)
plt.title('Box Plot: Population by Year')
plt.xlabel('Year')
plt.ylabel('Population (millions)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of Population by Age Groups
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='Population aged 0 to 14 years old (percentage)', data=df, label='0-14 years')
sns.lineplot(x='Year', y='Population aged 60+ years old (percentage)', data=df, label='60+ years')
plt.title('Population Distribution by Age Groups Over Years')
plt.xlabel('Year')
plt.ylabel('Percentage')
plt.legend()
plt.show()
```

# Explore the GDP dataset

```python
# Load GDP dataset
gdp_df = pd.read_excel('cleaned_gdp.xlsx')

# Display first five rows
gdp_df.head()
```

```python
# Summary statistics
gdp_df.describe()
```

```python
# Data types information
gdp_df.info()
```

```python
# Identify missing values
gdp_df.isnull().sum()
```

```python
# Z-Score outliers
z_scores = zscore(gdp_df.select_dtypes(include='number'))
outliers = (abs(z_scores) > 3).sum(axis=0)
outliers
```

```python
# Plot histograms
numeric_cols = ['GDP in current prices (millions of US dollars)', 
                'GDP per capita (US dollars)', 
                'GDP real rates of growth (percent)']

plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(gdp_df[col].dropna(), kde=True, bins=40)
    plt.title(col)
    
    # Set x-axis limits based on the distribution of each column
    if col == 'GDP in current prices (millions of US dollars)':
        plt.xlim(0, gdp_df[col].dropna().quantile(0.99))
    elif col == 'GDP per capita (US dollars)':
        plt.xlim(0, gdp_df[col].dropna().quantile(0.97))
    elif col == 'GDP real rates of growth (percent)':
        plt.xlim(gdp_df[col].min(), gdp_df[col].max())

plt.tight_layout()
plt.show()
```

```python
# Box plot for 'GDP real rates of growth (percent)' by year
plt.figure(figsize=(10, 6))
sns.boxplot(x='Year', y='GDP real rates of growth (percent)', data=gdp_df)
plt.title('Box Plot: GDP Growth Rate by Year')
plt.xlabel('Year')
plt.ylabel('GDP Growth Rate (%)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plot for average GDP per capita by year
avg_gdp_per_capita_by_year = gdp_df.groupby('Year')['GDP per capita (US dollars)'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Year', y='GDP per capita (US dollars)', data=avg_gdp_per_capita_by_year)
plt.title('Average GDP per Capita by Year')
plt.xlabel('Year')
plt.ylabel('GDP per capita (US dollars)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribution of GDP growth rates over years
plt.figure(figsize=(12, 6))
sns.lineplot(x='Year', y='GDP real rates of growth (percent)', data=gdp_df)
plt.title('GDP Growth Rates Over Years')
plt.xlabel('Year')
plt.ylabel('GDP Growth Rate (%)')
plt.show()
```
